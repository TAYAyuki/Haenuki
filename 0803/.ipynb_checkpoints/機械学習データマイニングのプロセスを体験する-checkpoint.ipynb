```python
import pandas as pd 
import numpy as np
from sklearn.tree import DecisionTreeClassifier as DT
```

```python
train = pd.read_csv("bank-additional-full.csv",delimiter=";")
```

```python
train.head()
```

```python
train.shape
```

```python
#目的変数を抜き出す
y=train["y"]
```

```python
#説明変数を抜き出す
trainx = train.iloc[:,0:20]
trainx.head()
```

```python
#ダミー変数化をおこなう
trainxd = pd.get_dummies(trainx)
#trainxd.info()
```

```python
#モデルの作成
clf3 = DT(max_depth=3 , min_samples_leaf=500)
clf3.fit(trainxd,y)
```

```python
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image
export_graphviz(clf3, out_file="tree_clf3.dot", feature_names=trainxd.columns, class_names=["0","1"], filled=True, rounded=True)
g = pydotplus.graph_from_dot_file(path="tree_clf3.dot")
Image(g.create_png())
```

```python
#Testデータを見てみましょう。
test = pd.read_csv("bank-additional.csv",delimiter=";")
```

```python
test.head()
```

```python
#説明変数を抜き出す
testx = test.iloc[:,0:20]
#ダミー変数化する
testxd = pd.get_dummies(testx)
#作成したモデルを使って予測を行う
pred =  clf3.predict_proba(testxd)
```

```python
#予測値を取り出す
p = pred[:,1]
p = pd.DataFrame(p, columns=["predict"])
#元のテストデータに予測値フィールドを追加する
test_p = pd.concat([test,p],axis=1)
test_p.to_csv("clf3_test_p.csv")
```

```python
test_p.head()
```
