```python
X = data_hokan.select_dtypes(exclude=['object'])
X = X[X.columns[X.columns != 'disease']]

y = data['disease']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
random_forest = RandomForestClassifier(max_depth=3, n_estimators=10, random_state=5)
random_forest.fit(X,y)
random_forest.score(X,y)
```
